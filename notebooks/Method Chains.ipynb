### Make keep() method return Clumper for chaining

```python
class Clumper:
    def __init__(self, blob):
        self.blob = blob

    def keep(self, func):
        return Clumper([d for d in self.blob if func(d)])
```

```python
(Clumper(poke_dict)
    .keep(lambda d: 'Grass' in d['type'])
    .keep(lambda d: d['hp'] < 60)
    .blob)
```

## *args

```python
class Clumper:
    def __init__(self, blob):
        self.blob = blob

    def keep(self, *funcs):
        data = self.blob
        for func in funcs:
            data = [d for d in data if func(d)]
        return Clumper(data)
```

```python
(Clumper(poke_dict)
    .keep(lambda d: 'Grass' in d['type'])
    .keep(lambda d: d['hp'] < 60)
    .blob)
```

```python
(Clumper(poke_dict)
    .keep(lambda d: 'Grass' in d['type'],
          lambda d: d['hp'] < 60)
    .blob)
```

## Head/Tails

```python
class Clumper:
    def __init__(self, blob):
        self.blob = blob

    def keep(self, *funcs):
        data = self.blob
        for func in funcs:
            data = [d for d in data if func(d)]
        return Clumper(data)

    def head(self, n):
        return Clumper([self.blob[i] for i in range(n)])

    def tail(self, n):
        return Clumper([self.blob[-i] for i in range(1, n+1)])
```

```python
(Clumper(poke_dict)
    .keep(lambda d: 'Grass' in d['type'],
          lambda d: d['hp'] < 60)
    .tail(2)
    .blob)
```

## Select

```python
class Clumper:
    def __init__(self, blob):
        self.blob = blob

    def keep(self, *funcs):
        data = self.blob
        for func in funcs:
            data = [d for d in data if func(d)]
        return Clumper(data)

    def head(self, n):
        return Clumper([self.blob[i] for i in range(n)])

    def tail(self, n):
        return Clumper([self.blob[-i] for i in range(1, n+1)])

    def select(self, *keys):
        return Clumper([{k: d[k] for k in keys} for d in self.blob])
```

```python
(Clumper(poke_dict)
    .keep(lambda d: 'Grass' in d['type'],
          lambda d: d['hp'] < 60)
    .tail(2)
    .select('name', 'hp')
    .blob)
```

## Mutate

```python
class Clumper:
    def __init__(self, blob):
        self.blob = blob

    def keep(self, *funcs):
        data = self.blob
        for func in funcs:
            data = [d for d in data if func(d)]
        return Clumper(data)

    def head(self, n):
        return Clumper([self.blob[i] for i in range(n)])

    def tail(self, n):
        return Clumper([self.blob[-i] for i in range(1, n+1)])

    def select(self, *keys):
        return Clumper([{k: d[k] for k in keys} for d in self.blob])

    def mutate(self, **kwargs):
      data = self.blob
      for key, func in kwargs.items():
          for i in range(len(data)):
              data[i][key] = func(data[i])
      return Clumper(data)
```

```python
(Clumper(poke_dict)
    .keep(lambda d: 'Grass' in d['type'],
          lambda d: d['hp'] < 60)
    .tail(2)
    .select('name', 'hp')
    .mutate(hp=lambda d: d['hp'] * 2,
            hp4=lambda d: d['hp'] * 4)
    .blob)
```

## Sort

```python
class Clumper:
    def __init__(self, blob):
        self.blob = blob

    def keep(self, *funcs):
        data = self.blob
        for func in funcs:
            data = [d for d in data if func(d)]
        return Clumper(data)

    def head(self, n):
        return Clumper([self.blob[i] for i in range(n)])

    def tail(self, n):
        return Clumper([self.blob[-i] for i in range(1, n+1)])

    def select(self, *keys):
        return Clumper([{k: d[k] for k in keys} for d in self.blob])

    def mutate(self, **kwargs):
        data = self.blob
        for key, func in kwargs.items():
            for i in range(len(data)):
                data[i][key] = func(data[i])
        return Clumper(data)

    def sort(self, key, reverse=False):
        return Clumper(sorted(self.blob, key=key, reverse=reverse))
```

```python
tuple_list = [(1, 2), (4, 3), (5, 10), (10, 2)]
```

```python
sorted(tuple_list, key=lambda t: t[1])
```

```python
(Clumper(poke_dict)
    .keep(lambda d: 'Grass' in d['type'],
          lambda d: d['hp'] < 60)
    .select('name', 'hp')
    .sort(lambda d: d['hp'], reverse=True)
    .blob)
```

```python
(Clumper(poke_dict)
  .keep(lambda d: 'Grass' in d['type'],
        lambda d: d['hp'] < 60)
  .mutate(ratio=lambda d: d['attack']/d['hp'])
  .select('name', 'ratio')
  .sort(lambda d: d['ratio'], reverse=True)
  .head(15)
  .blob)
```
